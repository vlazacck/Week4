```python
# Function to detect outliers using IQR
import logging 
import pandas as pd


logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("analysis.log"),
        logging.StreamHandler()
    ]
)

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers in critical columns
logging.info("Detecting outliers...")
sales_outliers = detect_outliers(train_cleaned, 'Sales')
customers_outliers = detect_outliers(train_cleaned, 'Customers')
competition_outliers = detect_outliers(store_data, 'CompetitionDistance')

# Summary of outliers
logging.info(f"Sales Outliers: {len(sales_outliers)}")
logging.info(f"Customers Outliers: {len(customers_outliers)}")
logging.info(f"Competition Distance Outliers: {len(competition_outliers)}")
```

```python
# Example: Capping outliers
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("test_analysis.log"),
        logging.StreamHandler()
    ]
)

logging.info("Capping outliers...")
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

# Apply capping
train_cleaned = cap_outliers(train_cleaned, 'Sales')
train_cleaned = cap_outliers(train_cleaned, 'Customers')
store_data = cap_outliers(store_data, 'CompetitionDistance')
logging.info("Outliers capped successfully.")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta


logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("analysis.log"),
        logging.StreamHandler()
    ]
)

logging.info(f"Seasonal Sales Summary ")

# Ensure the output directory exists
output_dir = "outputs"
os.makedirs(output_dir, exist_ok=True)

# Use existing merged_train_data
data = merged_train_data.copy()

# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Define holiday periods (Christmas and Easter for analysis)
christmas_period = [(12, 23), (12, 24), (12, 25), (12, 26)]  # Including Christmas Eve and Boxing Day
easter_period = [(4, 1), (4, 2), (4, 3), (4, 4)]  # Approximate Easter dates (will vary by year)

def seasonal_category(row):
    month = row['Date'].month
    day = row['Date'].day
    if (month, day) in christmas_period:
        return 'Christmas'
    elif (month, day) in easter_period:
        return 'Easter'
    else:
        return 'Non-Seasonal'

# Categorize days into seasonal categories
data['SeasonalCategory'] = data.apply(seasonal_category, axis=1)

# Group by SeasonalCategory and calculate mean sales
seasonal_sales = data.groupby('SeasonalCategory')['Sales'].mean().reset_index()
seasonal_sales = seasonal_sales.sort_values(by='Sales', ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(data=seasonal_sales, x='SeasonalCategory', y='Sales', palette='coolwarm')
plt.title('Average Sales During Seasonal Periods', fontsize=14)
plt.ylabel('Average Sales', fontsize=12)
plt.xlabel('Seasonal Category', fontsize=12)
plt.xticks(rotation=45)

# Save the plot
seasonal_sales_plot_path = os.path.join(output_dir, 'seasonal_sales_comparison.png')
plt.savefig(seasonal_sales_plot_path)
plt.show()

# Display the summary table in the notebook
print("Seasonal Sales Summary:\n")
print(seasonal_sales.to_string(index=False))

# Save the summary as a CSV file
seasonal_sales_summary_path = os.path.join(output_dir, 'seasonal_sales_summary.csv')
seasonal_sales.to_csv(seasonal_sales_summary_path, index=False)

print(f"Visualization saved to: {seasonal_sales_plot_path}")
print(f"Summary table saved as '{seasonal_sales_summary_path}' in outputs folder.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import logging

# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("analysis.log"),
        logging.StreamHandler()
    ]
)

logging.info("Script started: Weekday vs. Weekend Sales Analysis")

# Ensure output directory exists
output_dir = "outputs"
os.makedirs(output_dir, exist_ok=True)
logging.info(f"Output directory ensured: {output_dir}")

# Create a copy of the data for analysis
data = merged_train_data.copy()
data['Date'] = pd.to_datetime(data['Date'])
data['DayOfWeek'] = data['Date'].dt.dayofweek + 1  # Monday=1, Sunday=7
logging.info("Data preprocessing completed: Added 'Date' and 'DayOfWeek' columns")

# Analyze stores open on all weekdays
weekday_data = data[(data['DayOfWeek'] <= 5) & (data['Open'] == 1)]
all_weekday_stores = weekday_data.groupby('Store')['DayOfWeek'].nunique()
logging.info("Identified stores open on all weekdays")

# Identify stores open on all weekdays
all_weekday_stores = all_weekday_stores[all_weekday_stores == 5].index
logging.info(f"Stores open all weekdays: {len(all_weekday_stores)} stores found")

# Filter data for these stores
data_all_weekdays = data[data['Store'].isin(all_weekday_stores)]
data_all_weekdays['Weekend'] = data_all_weekdays['DayOfWeek'].apply(lambda x: 1 if x >= 6 else 0)
logging.info("Filtered data for stores open all weekdays and added 'Weekend' column")

# Analyze weekend vs weekday sales
weekend_sales = data_all_weekdays.groupby('Weekend')['Sales'].mean().reset_index()
weekend_sales['Type'] = weekend_sales['Weekend'].map({0: 'Weekday', 1: 'Weekend'})
logging.info("Calculated average sales for weekdays and weekends")

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(data=weekend_sales, x='Type', y='Sales', palette='viridis')
plt.title('Average Sales for Stores Open on All Weekdays', fontsize=14)
plt.xlabel('Type of Day', fontsize=12)
plt.ylabel('Average Sales', fontsize=12)

# Save the plot
plot_path = os.path.join(output_dir, 'weekday_weekend_sales_comparison.png')
plt.savefig(plot_path)
plt.show()
logging.info(f"Visualization saved to: {plot_path}")

# Display summary table
print("Sales for Stores Open All Weekdays:\n")
print(weekend_sales[['Type', 'Sales']].to_string(index=False))
logging.info("Displayed sales summary table")

# Save summary as CSV
summary_path = os.path.join(output_dir, 'weekday_weekend_sales_summary.csv')
weekend_sales[['Type', 'Sales']].to_csv(summary_path, index=False)
logging.info(f"Summary table saved as: {summary_path}")

logging.info("Script completed successfully")
```
